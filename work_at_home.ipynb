```python
import pandas as pd
df_lines = pd.read_csv('./data_line/20190124_bus_lines.csv')
```

```python
df_lines.head()
```

```python
line_names = df_lines['노선명'].unique()
len(line_names)
```

```python
for nm in line_names:
    if '6645' in nm:
        print(nm)
        
```

```python
df_le = df_lines[df_lines['노선명']=='6632']
df_le
```

```python
import folium
map = folium.Map(location=[df_le['Y좌표'].mean(), df_le['X좌표'].mean()], zoom_start=12, tiles='Stamen Toner')

for n in df_le.index[:]:
    # 팝업에 들어갈 텍스트를 지정해 줍니다.
    popup_name = df_le['정류소명'][n]
    # 브랜드명에 따라 아이콘 색상을 달리해서 찍어줍니다.
    icon_color = 'red'
        
    # folium.features.CircleMarker 오류가 날 경우 --> folium.vector_layer.CircleMarker
    folium.CircleMarker(
        location=[df_le['Y좌표'][n], df_le['X좌표'][n]],
        radius=3,
        popup=popup_name,
        color=icon_color,
        fill=True,
        fill_color=icon_color
    ).add_to(map)    

map
```

```python
df_le = df_le.drop(columns=['Unnamed: 0'])
```

```python
df_le.columns, df_le.index
```

```python
df_psngr.columns, df_psngr.index
```

```python
df_joined = df_psngr.merge(df_le, left_on="표준버스정류장ID", right_on="정류소ID")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20, 5))
geton = plt.bar(df_joined['순번_정류장명'], df_joined['8시승차총승객수'])
getoff = plt.bar(df_joined['순번_정류장명'], df_joined['8시하차총승객수'], bottom=df_joined['8시승차총승객수'])
plt.xticks(rotation=90)
plt.legend((geton[0], getoff[0]), ('BOARD', 'ALIGHT'))
plt.title('8시')
plt.show()
```
