# 노선의 정차순번 고려

```python
import pandas as pd
df_lines = pd.read_csv('./data_line/20190124_bus_lines.csv', encoding='cp949')
df_le = df_lines[df_lines['노선명']=='은평01']
df_le
```

```python
import folium
map = folium.Map(location=[df_le['Y좌표'].mean(), df_le['X좌표'].mean()], zoom_start=12, tiles='Stamen Toner')

for n in df_le.index[:]:
    # 팝업에 들어갈 텍스트를 지정해 줍니다.
    popup_name = df_le['정류소명'][n]
    # 브랜드명에 따라 아이콘 색상을 달리해서 찍어줍니다.
    icon_color = 'red'
        
    # folium.features.CircleMarker 오류가 날 경우 --> folium.vector_layer.CircleMarker
    folium.CircleMarker(
        location=[df_le['Y좌표'][n], df_le['X좌표'][n]],
        radius=3,
        popup=popup_name,
        color=icon_color,
        fill=True,
        fill_color=icon_color
    ).add_to(map)    

map
```

```python
df_joined = df_eunp01.merge(df_le, left_on="표준버스정류장ID", right_on="정류소ID")
df_joined['순번_정류장명'] = df_joined['순번'].astype(str) + "_" + df_joined['역명']

df_joined = df_joined.set_index('순번')
df_joined = df_joined.sort_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig,(ax1,ax2)= plt.subplots(nrows=2)
fig.set_size_inches(12,10)
sns.barplot(data=df_joined, y='순번_정류장명', x='8시승차총승객수', ax=ax1)
ax1.set(title="승차수")
sns.barplot(data=df_joined, y='순번_정류장명', x='8시하차총승객수', ax=ax2)
ax2.set(title="하차수")
```
